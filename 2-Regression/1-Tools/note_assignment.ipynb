```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets, linear_model, model_selection

X, y = datasets.load_linnerud(return_X_y=True, as_frame=False)
print(X.shape)
print(y.shape)

X = X[:, 1]
X = X.reshape((-1,1))
y = y[:, 1]
y = y.reshape((-1, 1))

X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y, test_size=0.50)

model = linear_model.LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)

plt.scatter(X_test, y_test,  color='black')
plt.plot(X_test, y_pred, color='red', linewidth=3)
plt.xlabel('Waistline in cm')
plt.ylabel('Number of situps accomplished')
plt.title('A Graph Plot Showing The Relationship Between Waistline And Number Of Situps Accomplished')
plt.show()
```
