# Playing with graphs

```python
import pandas as pd
import matplotlib.pyplot as plt

pumpkins = pd.read_csv('../data/US-pumpkins.csv')

# pumpkins = pumpkins[pumpkins['Variety']]
```

```python
columns_to_select = ['Variety', 'Low Price', 'High Price']
pumpkins = pumpkins.loc[:, columns_to_select]

pumpkins = pumpkins.dropna(subset='Variety')
pumpkins.isnull().sum()
```

```python
price = (pumpkins['Low Price'] + pumpkins['High Price']) / 2
new_pumpkins = pd.DataFrame({'Variety': pumpkins['Variety'], 'Low Price': pumpkins['Low Price'],'High Price': pumpkins['High Price'], 'Price': price})
```

```python
new_pumpkins.groupby(['Variety'])['Price'].mean().plot(kind='bar')
plt.ylabel("Pumpkin Price")
```
